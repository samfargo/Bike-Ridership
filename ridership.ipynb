 <font size="4">**Collect and Clean Data**</font>

Import modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import date
```

Bring in .csv file

```python
df = pd.read_csv('201906-bluebikes-tripdata.csv')
```

Preview the data. Most of the data cleaning will involve formatting the starttime column to properly show day of the week and hour.

```python
df[['starttime','tripduration']].head()
```

Create a new column containing 'starttime' converted to a datetime data type. This allows datetime operations to be performed on the column.

```python
df['starttime_datetime'] = pd.to_datetime(df['starttime'])
```

Use trips from 6/2/2019 to 6/29/2019 so each day of the week appears the same number of times.

```python
df = df[df['starttime_datetime']<date(2019, 6, 30)]
df = df[df['starttime_datetime']>date(2019, 6, 2)]
```

Create a new column containing the day name of the start of each trip (i.e. Monday).

```python
df['starttime_day'] = df['starttime_datetime'].dt.weekday_name;
```

Create a new column containing the hour of each trip's start time.

```python
df['starttime_hour'] = df['starttime_datetime'].dt.hour;
```

Create a new dataframe to pivot the data by day and hour, then get the count of trips.

```python
df_pivot = pd.pivot_table(df[['starttime_day', 'starttime_hour','tripduration']], index=['starttime_day', 'starttime_hour'], aggfunc='count')
```

Place days on the x-axis and hour on the y-axis.

```python
trips = df_pivot.unstack(level=0)
```

Reindex so day of the week displays in chronological order.

```python
trips = trips.reindex(labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], axis=1, level=1)
```

 <font size="4">**Plot the Data**</font>

Create axis tick labels.

```python
am = []
for hour in range(1, 12):
    hour_list = str(hour) + "am"
    am.append(hour_list)
```

```python
pm = []
for hour in range(1, 12):
    hour_list = str(hour) + "pm"
    pm.append(hour_list)
```

```python
hour_list = ["12am"] + am + ["12pm"] + pm
```

```python
days = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
```

Plot heatmap of trip count by day & hour.

```python
f, ax = plt.subplots(figsize=(10, 15))
ax = sns.heatmap(trips, annot=True, fmt="d", ax=ax, xticklabels=days, yticklabels=hour_list)
ax.axes.set_title("Boston BlueBike Trip Count by Day & Hour (June 2019)", fontsize=14)
ax.set(xlabel='Day of Week', ylabel='Starting Hour of Ride');
plt.yticks(rotation=0)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5) #slightly shifts the plot to align the y-axis properly
plt.show()
```

 <font size="4">**Interpretation**</font>

From Monday to Friday, there is a clear spike in trips taken during standard commuting hours, spanning 7am - 9am and 4pm - 7pm.

On Saturday and Sunday, trips taken are primarily concentrated between typical lunch and dinner hours.

As expected, there are relatively very few trips taken throughout the night.
